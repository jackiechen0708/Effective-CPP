```python
# load data set
import sqlite3
import pandas as pd
conn = sqlite3.connect("house.db")
data = pd.read_sql_query("SELECT * FROM houseprice;",conn)
```

```python
# check the database
data.describe()
```

```python
data = data.fillna(0)
data.corr()
```

```python
# load data
#data = data.reset_index(drop=True)

# features = ['bedrooms', 'bathrooms', 'sqft_living', 'sqft_lot', 'floors','waterfront', 'view', 'condition',  'sqft_above',
#    'sqft_basement',  'sqft_living15']

features = [ 'sqft_living','overall_grade','sqft_above']

print(len(data))
train = data.iloc[0:17290]
train_x = train[features].values
train_y = train['price'].values


test = data.iloc[17291:18371]
test_x = test[features].values
test_y = test['price'].values
```

```python
# create model
from sklearn.linear_model  import LinearRegression
# Create linear regression object
regr = LinearRegression()

# Train the model using the training sets
regr.fit(train_x, train_y)

# Make predictions using the testing set
test_y_pred = regr.predict(test_x)

```

```python
# predict
print("result")
print(test_y_pred)
# The coefficients
print('Coefficients: \n', regr.coef_)
```

```python
# show result
import matplotlib
import matplotlib.pyplot as plot
%matplotlib inline
l1, = plot.plot(test_y,label='target')
l2,=plot.plot(test_y_pred,label = 'predict')
plot.legend(handles=[l1, l2])
plot.show()
```
